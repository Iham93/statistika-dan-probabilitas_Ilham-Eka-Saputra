```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Parameter distribusi
mean = 5
std_dev = 2
sample_size = 1000

# Bangkitkan data waktu pengiriman
delivery_times = np.random.normal(mean, std_dev, sample_size)

# Buat DataFrame
df = pd.DataFrame(delivery_times, columns=['Waktu Pengiriman'])

# Visualisasi distribusi waktu pengiriman
plt.figure(figsize=(10, 6))
sns.histplot(df['Waktu Pengiriman'], kde=True)
plt.title('Distribusi Waktu Pengiriman Paket')
plt.xlabel('Waktu Pengiriman (hari)')
plt.ylabel('Frekuensi')
plt.show()

# Hitung probabilitas
prob_kurang_dari_3 = np.sum(df['Waktu Pengiriman'] < 3) / sample_size
print(f"Probabilitas waktu pengiriman kurang dari 3 hari: {prob_kurang_dari_3:.2f}")

prob_antara_4_dan_7 = np.sum((df['Waktu Pengiriman'] >= 4) & (df['Waktu Pengiriman'] <= 7)) / sample_size
print(f"Probabilitas waktu pengiriman antara 4 dan 7 hari: {prob_antara_4_dan_7:.2f}")

# Visualisasi probabilitas kumulatif
plt.figure(figsize=(10, 6))
sns.ecdfplot(df['Waktu Pengiriman'])
plt.title('Probabilitas Kumulatif Waktu Pengiriman')
plt.xlabel('Waktu Pengiriman (hari)')
plt.ylabel('Probabilitas Kumulatif')
plt.show()

# Analisis statistik deskriptif
print(df['Waktu Pengiriman'].describe())
```
